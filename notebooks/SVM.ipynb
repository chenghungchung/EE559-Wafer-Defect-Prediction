```python
import pickle

with open("../data/preprocessed_data_vae_enhanced.pkl", "rb") as f:
    preprocessed_data = pickle.load(f)

X_train_scaled = preprocessed_data["X_train"]
X_val_scaled = preprocessed_data["X_val"]
X_test_scaled = preprocessed_data["X_test"]
y_train = preprocessed_data["y_train"]
y_val = preprocessed_data["y_val"]
y_test = preprocessed_data["y_test"]

print(X_train_scaled.shape)
print(X_val_scaled.shape)
print(X_test_scaled.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

RANDOM_SEED = 559

svm_baseline = SVC(kernel="rbf", random_state=RANDOM_SEED)
svm_baseline.fit(X_train_scaled, y_train)

y_val_pred = svm_baseline.predict(X_val_scaled)

print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Precision:", precision_score(y_val, y_val_pred, pos_label=1))
print("Validation Recall:", recall_score(y_val, y_val_pred, pos_label=1))
print("Validation F1-score:", f1_score(y_val, y_val_pred, pos_label=1))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

RANDOM_SEED = 559

svm_weighted = SVC(kernel="rbf", class_weight="balanced", random_state=RANDOM_SEED)
svm_weighted.fit(X_train_scaled, y_train)

y_val_pred = svm_weighted.predict(X_val_scaled)

print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

best_model = None
best_score = -1
best_setting = None

c_list = [0.01, 0.05, 0.1, 0.2, 0.5, 1]
weight_list = [
    "balanced",
    {-1: 1, 1: 5},
    {-1: 1, 1: 10},
    {-1: 1, 1: 15},
    {-1: 1, 1: 20},
]

results = []

for c in c_list:
    for w in weight_list:
        svm_model = SVC(
            kernel="linear",
            C=c,
            class_weight=w,
            random_state=RANDOM_SEED
        )

        svm_model.fit(X_train_scaled, y_train)
        y_val_pred = svm_model.predict(X_val_scaled)

        acc = accuracy_score(y_val, y_val_pred)
        prec = precision_score(y_val, y_val_pred, pos_label=1, zero_division=0)
        rec = recall_score(y_val, y_val_pred, pos_label=1, zero_division=0)
        f1 = f1_score(y_val, y_val_pred, pos_label=1, zero_division=0)
        cm = confusion_matrix(y_val, y_val_pred)

        results.append({
            "C": c,
            "class_weight": w,
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1": f1,
            "confusion_matrix": cm
        })

        print(f"\nLinear SVM | C={c} | class_weight={w}")
        print("Accuracy:", acc)
        print("Precision:", prec)
        print("Recall:", rec)
        print("F1-score:", f1)
        print("Confusion Matrix:\n", cm)

        score = (rec, f1)

        if score > best_score if isinstance(best_score, tuple) else (rec, f1) > (-1, -1):
            best_score = score
            best_model = svm_model
            best_setting = {
                "C": c,
                "class_weight": w,
                "accuracy": acc,
                "precision": prec,
                "recall": rec,
                "f1": f1,
                "confusion_matrix": cm
            }

print("\n" + "="*50)
print("Best validation setting:")
print("C:", best_setting["C"])
print("class_weight:", best_setting["class_weight"])
print("Accuracy:", best_setting["accuracy"])
print("Precision:", best_setting["precision"])
print("Recall:", best_setting["recall"])
print("F1-score:", best_setting["f1"])
print("Confusion Matrix:\n", best_setting["confusion_matrix"])
```
