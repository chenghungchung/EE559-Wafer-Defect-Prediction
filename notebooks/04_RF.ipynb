```python
import pickle
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

RANDOM_SEED = 559

with open("../data/preprocessed_data.pkl", "rb") as f:
    preprocessed_data = pickle.load(f)

X_train = preprocessed_data["X_train"]
X_val = preprocessed_data["X_val"]
X_test = preprocessed_data["X_test"]
y_train = preprocessed_data["y_train"]
y_val = preprocessed_data["y_val"]
y_test = preprocessed_data["y_test"]

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)

print("y_train unique:", np.unique(y_train))
print("y_val unique:", np.unique(y_val))
print("y_test unique:", np.unique(y_test))
```

```python
rf_baseline = RandomForestClassifier(
    n_estimators=200,
    random_state=RANDOM_SEED,
    n_jobs=-1
)

rf_baseline.fit(X_train, y_train)
y_val_pred = rf_baseline.predict(X_val)

print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
rf_weighted = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=RANDOM_SEED,
    n_jobs=-1
)

rf_weighted.fit(X_train, y_train)
y_val_pred = rf_weighted.predict(X_val)

print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Validation F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
for w in [3, 5, 10, 15]:
    rf_model = RandomForestClassifier(
        n_estimators=200,
        class_weight={-1: 1, 1: w},
        random_state=RANDOM_SEED,
        n_jobs=-1
    )

    rf_model.fit(X_train, y_train)
    y_val_pred = rf_model.predict(X_val)

    print(f"\nClass weight {{-1:1, 1:{w}}}")
    print("Accuracy:", accuracy_score(y_val, y_val_pred))
    print("Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
    print("Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
    print("F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
    print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
for n_estimators in [100, 200, 400]:
    for max_depth in [5, 10, None]:
        rf_model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            class_weight="balanced",
            random_state=RANDOM_SEED,
            n_jobs=-1
        )

        rf_model.fit(X_train, y_train)
        y_val_pred = rf_model.predict(X_val)

        print(f"\nRandom Forest, n_estimators={n_estimators}, max_depth={max_depth}")
        print("Accuracy:", accuracy_score(y_val, y_val_pred))
        print("Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
for min_samples_leaf in [1, 2, 4]:
    for min_samples_split in [2, 5, 10]:
        rf_model = RandomForestClassifier(
            n_estimators=200,
            max_depth=10,
            min_samples_leaf=min_samples_leaf,
            min_samples_split=min_samples_split,
            class_weight="balanced",
            random_state=RANDOM_SEED,
            n_jobs=-1
        )

        rf_model.fit(X_train, y_train)
        y_val_pred = rf_model.predict(X_val)

        print(f"\nRandom Forest, min_samples_leaf={min_samples_leaf}, min_samples_split={min_samples_split}")
        print("Accuracy:", accuracy_score(y_val, y_val_pred))
        print("Precision:", precision_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("Recall:", recall_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("F1-score:", f1_score(y_val, y_val_pred, pos_label=1, zero_division=0))
        print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

```python
best_rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=1,
    min_samples_split=2,
    class_weight="balanced",
    random_state=RANDOM_SEED,
    n_jobs=-1
)

best_rf.fit(X_train, y_train)
y_test_pred = best_rf.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Test Precision:", precision_score(y_test, y_test_pred, pos_label=1, zero_division=0))
print("Test Recall:", recall_score(y_test, y_test_pred, pos_label=1, zero_division=0))
print("Test F1-score:", f1_score(y_test, y_test_pred, pos_label=1, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
```

```python
feature_importance = pd.DataFrame({
    "feature": X_train.columns,
    "importance": best_rf.feature_importances_
}).sort_values(by="importance", ascending=False).reset_index(drop=True)

print("=== Top 20 Important Features ===")
display(feature_importance.head(20))
```
